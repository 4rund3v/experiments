#### Types of Inheritance - Vertical and Horizontal
The above examples represent vertical inheritance, python also supports horizontal inheritance(Multiple Inheritance), by allowing a subclass to inherit more than one base class at a time, this way a calssa can obtain behaviors from many various classes without having to go several levels deep.
This is particularly well suited for applications where your classes share common behaviours, but where they're not otherwise related  to each other in a hierarchial manner. 
MultipleInheritance is often called to apply support classes : also called as mixins.
Mixin classes don't provide full functionality on their own; they instead supply just a smalll add-on feature that could be useful on a wide range of classes.
Typically, a vertical hierarchy will provide most of the functionality with mixins adding some extras where necessary.
Because of the potentially large number of classes involved, it becomes even more important to fully understand how Python decides which class is used for which attribute and method that was accessed. 

##### Method Resolution Order 
 Given a class hierarchy, python needs to determine which class to use when attempting to access an attribute by name.
 In the vertical inheritance approach: the class you are working with will be the first in line followed by its base class, followed by its base class and its base class and so on until its at the root (object ) class.
 Example Below

```python
class Book():
    def __init__(self,title):
        self.title = title
        self.page = 1
        self.author  = ""
    
    def read(self):
        return "Currently reading the book : {} and on page : {}".format(self.title, self.page)
    
    def bookmark(self, page):
        self.page = page

class Novel(Book):
    def read(self):
        return "Enjoying reading [{}] on page [{}]".format(self.title, self.page)

class Mystery(Novel):
    def read(self):
        return "This book is damn scary!! [{}] and im just on : {}".format(self.title, self.page)
```

```python
python_book = Book("Pro Python")
python_book.read()
```

```python
harry_potter = Novel("HarryPotter and the work from home")
harry_potter.read()
```

```python
sherlock = Mystery("Sherlock Holmes")
sherlock.read()
```

In the horizontal inheritance approach : python goes from left to right in the order of inheritance, for each of the supplied base classes so its purely a horizaontal approach.
Example Below

```python
class Product():
    def purchase(self):
        print("You are purchasing a product!")

class BookProduct(Book, Product):
    pass

class NovelProduct(Novel, Product):
    def purchase(self):
        print("Have fun purchasing a novel. Worth it!!")

class MysteryProduct(Mystery, Product):
    def purchase(self):
        print("Really ?? You are purchasing a thriller !!")
```

The first namespace that python looks for is the objects namespace followed by the class's namespace
only then does it look into the inherited classes.
But when there are mutliple inheritences, pyhton does not follow depth first or breadth first approach, it uses an algorithm called as C3, in escense the c3 function will perform the necessary linearization , reducing the inheritance tree into a single list. 

```python
import itertools
def C3(cls, *mro_lists):
    # Make a copy so we don't change existing content
    mro_lists = [list(mro_list[:]) for mro_list in mro_lists]
    # Set up the new MRO with the class itself
    mro = [cls]
    while True:
        # Reset for the next round of tests
        candidate_found = False
        for mro_list in mro_lists:
            if not len(mro_list):
                # Any empty lists are of no use to the algorithm.
                continue
            # Get the first item as a potential candidate for the MRO.
            candidate = mro_list[0]
            if candidate_found:
                # Candidates promoted to the MRO are no longer of use.
                if candidate in mro:
                    mro_list.pop(0)
                # Don't bother checking any more candidates if one was found.
                continue
                
            if candidate in itertools.chain(*(x[1:] for x in mro_lists)) :
                # The candidate was found in an invalid position, so we
                # move on to the next MRO list to get a new candidate.
                continue
            else:
                # The candidate is valid and should be promoted to the MRO.
                mro.append(candidate)
                mro_list.pop(0)
                candidate_found = True
        if not sum(len(mro_list) for mro_list in mro_lists):
            # There are no MROs to cycle through, so we're all done.
            break
        if not candidate_found:
            # No valid candidate was available, so we have to bail out.
            break
        raise TypeError("Inconsistent MRO")
    return mro
# -> [Parent, Nextparent, object]
```

```python
class Book():
    def __init__(self, title):
        self.title = title
        self.page = 1
        self.author  = ""
    
    def read(self):
        return "Currently reading the book : {} and on page : {}".format(self.title, self.page)
    
    def bookmark(self, page):
        self.page = page

class Novel(Book):
    def __init__(self, title):
        super(Book, self).__init__(title)
        
    def read(self):
        return "Enjoying reading [{}] on page [{}]".format(self.title, self.page)

class Mystery(Novel):
    def __init__(self, title):
        super(Novel, self).__init__(title)
    def read(self):
        return "This book is damn scary!! [{}] and im just on : {}".format(self.title, self.page)
```

```python
sherlock = Mystery("Sherlock Holmes")
sherlock.read()
```
